```python
dfsales.drop(['Status', 'unnamed1'], axis=1, inplace=True)
dfsales.dropna(inplace=True)
dfsales.info()
```

```python
dfsales.rename(columns={'Marital_Status': 'Marital'}, inplace=True)
dfsales['Marital'] = dfsales['Marital'].map({0: 'Single', 1: 'Married'})
dfsales.head()
```

```python
dfsales['Gender'] = dfsales['Gender'].map({'F': 'Female', 'M': 'Male'})
dfsales.head()
```

```python
plt.figure(facecolor = 'orange', figsize = (10,9))
plt.subplot(2,2,1)
gender_counts = dfsales['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%0.1f%%', colors=['red', 'lightskyblue'])
plt.title('Gender Ratio')

plt.subplot(2,2,2)
count=sns.countplot(x='Gender',data=dfsales,width=0.5,palette='magma')
plt.title('Gender Count')
for bars in count.containers:
    count.bar_label(bars)
plt.subplot(2,2,3)
count=sns.countplot(x='Marital',data=dfsales,width=0.5,palette='Set2')
plt.title('Maritial Status Count')
for bars in count.containers:
    count.bar_label(bars)

plt.subplot(2,2,4)
gender_counts = dfsales['Marital'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%0.1f%%', colors=['brown', 'blue'])
plt.title('Maritial Ratio')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# ------------------ Helper Functions ------------------

def pie_chart(data, column, title, colors, position):
    plt.subplot(2, 2, position)
    counts = data[column].value_counts()
    plt.pie(
        counts,
        labels=counts.index,
        autopct='%1.1f%%',
        colors=colors,
        startangle=90,
        wedgeprops={'edgecolor': 'white'}
    )
    plt.title(title, fontsize=12, fontweight='bold')


def bar_chart(data, column, title, palette, position):
    plt.subplot(2, 2, position)
    ax = sns.countplot(
        x=column,
        data=data,
        palette=palette,
        width=0.6
    )
    plt.title(title, fontsize=12, fontweight='bold')

    # Add labels on bars
    for container in ax.containers:
        ax.bar_label(container, padding=2)

# ------------------ Visualization ------------------

plt.figure(figsize=(12, 10), facecolor='orange')

pie_chart(dfsales, 'Gender', 'Gender Ratio', ['#FF6F61', '#6BAED6'], 1)
bar_chart(dfsales, 'Gender', 'Gender Count', 'magma', 2)

bar_chart(dfsales, 'Marital', 'Marital Status Count', 'Set2', 3)
pie_chart(dfsales, 'Marital', 'Marital Ratio', ['#8C564B', '#1F77B4'], 4)

plt.tight_layout(pad=3)
plt.show()
```

```python
fig=plt.figure(figsize=(20, 15), facecolor='lightgray')

# 1. Gender vs Amount
plt.subplot(2, 3, 1)
gender_amount = dfsales.groupby('Gender')['Amount'].sum().reset_index()
bar=sns.barplot(x='Gender', y='Amount', data=gender_amount, palette='pastel')
plt.title('Total Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
for i in bar.containers:
  bar.bar_label(i ,fontsize=11,fmt='%d')

# 2. Age Group vs Amount
plt.subplot(2, 3, 2)
age_group_amount = dfsales.groupby('Age Group')['Amount'].sum().reset_index()
bar=sns.barplot(x='Age Group', y='Amount', data=age_group_amount, palette='viridis')
plt.title('Total Amount by Age Group')
for i in bar.containers:
  bar.bar_label(i ,fontsize=11,fmt='%d')
plt.xlabel('Age Group')
plt.ylabel('Total Amount')

# 3. State vs Amount
plt.subplot(2, 3, 3)
state_amount = dfsales.groupby('State')['Amount'].sum().sort_values(ascending=False).reset_index()
bar=sns.barplot(y='State', x='Amount', data=state_amount.head(10), palette='magma') # Showing top 10 states for readability
plt.title('Top 10 States by Total Amount')
for i in bar.containers:
  bar.bar_label(i ,fontsize=11,fmt='%d')
plt.xlabel('Total Amount')
plt.ylabel('State')

# 4. Occupation vs Amount
plt.subplot(2, 3, 4)
occupation_amount = dfsales.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()
bar=sns.barplot(y='Occupation', x='Amount', data=occupation_amount.head(10), palette='cividis') # Showing top 10 occupations
plt.title('Top 10 Occupations by Total Amount')
for i in bar.containers:
  bar.bar_label(i ,fontsize=11,fmt='%d')
plt.xlabel('Total Amount')
plt.ylabel('Occupation')

# 5. Product Category vs Amount
plt.subplot(2, 3, 5)
product_amount = dfsales.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).reset_index()
bar=sns.barplot(y='Product_Category', x='Amount', data=product_amount.head(10), palette='plasma') # Showing top 10 product categories
plt.title('Top 10 Product Categories by Total Amount')
for i in bar.containers:
  bar.bar_label(i ,fontsize=11,fmt='%d')
plt.xlabel('Total Amount')
plt.ylabel('Product Category')

# 6. Marital vs Amount
plt.subplot(2, 3, 6)
marital_amount = dfsales.groupby('Marital')['Amount'].sum().reset_index()
bar=sns.barplot(x='Marital', y='Amount', data=marital_amount, palette='coolwarm')
plt.title('Total Amount by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Amount')
for i in bar.containers:
  bar.bar_label(i ,fontsize=11,fmt='%d')
fig.savefig('SalesProject.jpg')
plt.tight_layout()
plt.show()
```
